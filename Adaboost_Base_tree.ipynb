# 电信套餐匹配——AdaBoost_Base_LR——训练

```python
# 首先 import 必要的模块
import pandas as pd 
import numpy as np

from sklearn.model_selection import GridSearchCV

import matplotlib.pyplot as plt
%matplotlib inline
```

## 读取数据

```python
# 读取数据,原始特征MinMaxScale之后的结果
# path to where the data lies
dpath = './input/'
train = pd.read_csv(dpath +"train_data.csv")
test = pd.read_csv(dpath +"test_data.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
#首先把y_target取出来，并进行转化
class_mapping={89950166:0,
               89950167:1,
               89950168:2,
               90063345:3,
               90109916:4,
               90155946:5,
               99999825:6,
               99999826:7,
               99999827:8,
               99999828:9,
               99999830:10}
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_y_train = LabelEncoder()
train['current_service']= labelencoder_y_train.fit_transform(train['current_service'])
```

```python
y_train=train.pop('current_service')
train_user_id=train.pop('user_id')
X_train = train.drop([ "data_type"], axis=1)
```

```python
test_user_id=test.pop('user_id')
X_test = test.drop([ "data_type","current_service"], axis=1)
```

## AdaBoost_Base_tree

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_gaussian_quantiles
bdt = AdaBoostClassifier(DecisionTreeClassifier(class_weight="balanced",max_depth=80, min_samples_split=20, min_samples_leaf=8),
                         algorithm="SAMME",
                         n_estimators=100)
bdt.fit(X_train, y_train)
```

```python
import pickle as pk
pk.dump(bdt, open("Adaboost_base_tree.pkl", 'wb'))
```

```python
y_pred=bdt.predict(X_train)
print(y_pred)
```

```python
y_true=y_train
from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred))
```
