## 2. feat编码：log(x+1)
原始特征feat_x看起来像计数特征，取log运算更接近人对数字的敏感度，更适合线性模型。
同时也可以降低长维分布中大数值的影响，减弱长维分布的长尾性。

```python
all_df.describe()
```

```python
#对online_time进行log(x+1)处理
all_df['online_time'] = np.log1p(all_df['online_time'])
all_df.rename(columns={'online_time':'online_time_log'},inplace = True)
all_df.describe()
```

```python
#对total_fee_1,total_fee_2,total_fee_3,total_fee_4进行log(x+1)处理
#对total_fee_1
all_df['total_fee_1']=all_df['total_fee_1'].apply(lambda x:abs(x))
all_df['total_fee_1'] = np.log1p(all_df['total_fee_1'])
all_df.rename(columns={'total_fee_1':'total_fee_1_log'},inplace = True)

#对total_fee_2
all_df['total_fee_2']=all_df['total_fee_2'].apply(lambda x:abs(x))
all_df['total_fee_2'] = np.log1p(all_df['total_fee_2'])
all_df.rename(columns={'total_fee_2':'total_fee_2_log'},inplace = True)

#对total_fee_3
all_df['total_fee_3']=all_df['total_fee_3'].apply(lambda x:abs(x))
all_df['total_fee_3'] = np.log1p(all_df['total_fee_3'])
all_df.rename(columns={'total_fee_3':'total_fee_3_log'},inplace = True)

#对total_fee_4
all_df['total_fee_4']=all_df['total_fee_4'].apply(lambda x:abs(x))
all_df['total_fee_4'] = np.log1p(all_df['total_fee_4'])
all_df.rename(columns={'total_fee_4':'total_fee_4_log'},inplace = True)

all_df.describe()
```

```python
#对month_traffic进行log(x+1)处理
all_df['month_traffic'] = np.log1p(all_df['month_traffic'])
all_df.rename(columns={'month_traffic':'month_traffic'},inplace = True)
all_df.describe()
```

```python
#对pay_times进行log(x+1)处理
all_df['pay_times'] = np.log1p(all_df['pay_times'])
all_df.rename(columns={'pay_times':'pay_times_log'},inplace = True)
all_df.describe()
```

```python
#对pay_num进行log(x+1)处理
all_df['pay_num'] = np.log1p(all_df['pay_num'])
all_df.rename(columns={'pay_num':'pay_num_log'},inplace = True)
all_df.describe()
```

```python
#对last_month_traffic进行log(x+1)处理
all_df['last_month_traffic'] = np.log1p(all_df['last_month_traffic'])
all_df.rename(columns={'last_month_traffic':'last_month_traffic_log'},inplace = True)
all_df.describe()
```

```python
#对local_trafffic_month进行log(x+1)处理
all_df['local_trafffic_month'] = np.log1p(all_df['local_trafffic_month'])
all_df.rename(columns={'local_trafffic_month':'local_trafffic_month_log'},inplace = True)
all_df.describe()
```

```python
#对local_caller_time进行log(x+1)处理
all_df['local_caller_time'] = np.log1p(all_df['local_caller_time'])
all_df.rename(columns={'local_caller_time':'local_caller_time_log'},inplace = True)
all_df.describe()
```

```python
#对service1_caller_time进行log(x+1)处理
all_df['service1_caller_time'] = np.log1p(all_df['service1_caller_time'])
all_df.rename(columns={'service1_caller_time':'service1_caller_time_log'},inplace = True)
all_df.describe()
```
